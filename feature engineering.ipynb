```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
data_folder = os.path.join(cwd, "sales_forecasting_data")

df = pd.read_csv(os.path.join(data_folder, "clean_train.csv"))
```

```python
df.sort_values(['store_nbr', 'item_nbr', 'date'], inplace=True)
```

```python
df
```

```python
# Add dow and month
df['date'] = pd.to_datetime(df['date'])

df['dow'] = df['date'].dt.dayofweek
df['month'] = df['date'].dt.month
```

```python
df = df.drop('holiday', axis=1)
```

```python
df
```

```python
df['is_weekend'] = df['dow'].isin([5, 6]).astype(int)
```

```python
df = df.drop('dow', axis=1)
```

```python
df
```

```python
# Add sale lag
df['sale_lag_1'] = df.groupby(['store_nbr', 'item_nbr'])['unit_sales'].shift(1)
df['sale_lag_7'] = df.groupby(['store_nbr', 'item_nbr'])['unit_sales'].shift(7)
```

```python
# Add promotion lag
df['promo_lag_1'] = df.groupby(['store_nbr', 'item_nbr'])['onpromotion'].shift(1)
df['promo_lag_7'] = df.groupby(['store_nbr', 'item_nbr'])['onpromotion'].shift(7)
```

```python
df
```

```python
lag_cols = ['sale_lag_1', 'sale_lag_7', 'promo_lag_1', 'promo_lag_7']

for col in lag_cols:
    df[col] = df[col].fillna(0)

df.dropna(subset=lag_cols, inplace=True)
```

```python
df
```

```python
df_train = df[(df['date'] >= '2017-01-01') & (df['date'] < '2017-06-19')]
df_validation = df[(df['date'] >= '2017-06-20') & (df['date'] < '2017-07-19')]
df_test = df[(df['date'] >= '2017-07-20')]
```

```python
df_train.to_csv(os.path.join(data_folder, "engineered_train.csv"), index=False)
df_validation.to_csv(os.path.join(data_folder, "engineered_validation.csv"), index=False)
df_test.to_csv(os.path.join(data_folder, "engineered_test.csv"), index=False)
```
