```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
data_folder = os.path.join(cwd, "sales_forecasting_data")

train = pd.read_csv(os.path.join(data_folder, "train_sample.csv"), parse_dates=['date'])
stores = pd.read_csv(os.path.join(data_folder, "stores.csv"))
transactions = pd.read_csv(os.path.join(data_folder, "transactions.csv"), parse_dates=['date'])
oil = pd.read_csv(os.path.join(data_folder, "oil.csv"), parse_dates=['date'])
items = pd.read_csv(os.path.join(data_folder, "items.csv"))
holidays_events = pd.read_csv(os.path.join(data_folder, "holidays_events.csv"), parse_dates=['date'])
```

```python
# Drop transferred holiday
holidays_events = holidays_events[holidays_events["transferred"] == False]
holidays_events
```

```python
df = train.merge(items, on='item_nbr', how='left')

df = df.merge(stores, on='store_nbr', how='left')

df = df.merge(transactions, on=['date', 'store_nbr'], how='left')

df = df.merge(oil, on='date', how='left')

df = df.merge(holidays_events[['date', 'type']], on='date', how='left')
```

```python
df
```

```python
df['holiday'] = df['type_y'] == 'Holiday'
df = df.drop(columns=['type_y'])
```

```python
df = df.sort_values(by='date')
df
```

```python
# Handle missing oil price: backward fill
df['dcoilwtico'] = df['dcoilwtico'].fillna(method='bfill')
```

```python
df
```

```python
df['transactions'].value_counts()
```

```python
# Handle nan for promotions and transactions
df['onpromotion'] = df['onpromotion'].fillna(False)
df['transactions'] = df['transactions'].fillna(0)
```

```python
df
```

```python
df.to_csv(os.path.join(data_folder, "clean_train.csv"), index=False)
```
